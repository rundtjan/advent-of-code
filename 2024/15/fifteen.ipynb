```python
from functools import reduce

filename = 'input.txt'

with open(filename) as f:
    inputz = [list(line.strip()) for line in f.readlines()]

mapz = list(filter(lambda x: x is not None, [x if '#' in x else None for x in inputz]))

movez = reduce(lambda x, y: x+y, list(filter(lambda x: x is not None, [x if '#' not in x else None for x in inputz])))
```

```python
robo_pos = [0,0]

for y, row in enumerate(mapz):
    for x, elem in enumerate(row):
        if elem == '@':
            robo_pos = [y,x]
robo_pos
```

```python
def move_if_possible(position, direction, is_robo=False):
    if mapz[position[0] + direction[0]][position[1] + direction[1]] == '#':
        what_to_do = 'STOP'
    elif mapz[position[0] + direction[0]][position[1] + direction[1]] == '.':
        what_to_do =  'GO'
    else:
        what_to_do = move_if_possible((position[0] + direction[0], position[1] + direction[1]), direction)
    if what_to_do == 'STOP':
        return what_to_do
    else:
        mapz[position[0] + direction[0]][position[1] + direction[1]] = mapz[position[0]][position[1]]
        mapz[position[0]][position[1]]  = '.'
        if is_robo:
            robo_pos[0] = position[0] + direction[0]
            robo_pos[1] = position[1] + direction[1]
        return what_to_do
```

```python
directions = {'>': (0, 1), '<': (0, -1), '^': (-1, 0), 'v': (1, 0)}
```

```python
def print_map():
    for row in mapz:
        print(row)

print_map()
```

```python
for move in movez:
    move_if_possible(robo_pos, directions[move], is_robo=True)
```

```python
print_map()
```
