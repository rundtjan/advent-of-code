```python
from typing import Literal

class GuardTracker:
    map_width: int = 0
    map_height: int = 0
    obstacles_y_axis: dict = {}
    obstacles_x_axis: dict = {}
    guard_position: dict = {'y':0, 'x':0}
    guard_direction: Literal['up', 'right', 'down', 'left'] = 'up'
    file_name: str
    map: list[list[str]]
    loop_possibility_count = 0
    
    def __init__(self, file_name):
        self.file_name = file_name
        with open(file_name) as f:
            map_rows_raw = list(f.readlines())
            self.map = [x[:len(x)-1] for x in [list(x) for x in map_rows_raw]]
            self.map_width = len(self.map[0])
            self.map_height = len(self.map)
            break_flag = False
            for y, row in enumerate(self.map):
                if break_flag:
                    break
                for x, elem in enumerate(row):
                    if elem == '^':
                        self.guard_position['x'] = x
                        self.guard_position['y'] = y
                        self.map[y][x] = 'X'
                        break
                        
    
    def _draw_map(self):
        self.map[self.guard_position.get('y')][self.guard_position.get('x')] = '&'
        for line in self.map:
            print("".join(line))
        self.map[self.guard_position.get('y')][self.guard_position.get('x')] = 'X'
            
    
    def _move_guard(self):
        if self.guard_direction == 'up':
            if self.guard_position.get('y') == 0:
                return True
            elif self.map[self.guard_position.get('y') - 1][self.guard_position.get('x')] == '#':
                self.guard_direction = 'right'
                return False
            else:
                self.guard_position['y'] -= 1
                self.map[self.guard_position.get('y')][self.guard_position.get('x')] = 'X'
                return False
        elif self.guard_direction == 'down':
            if self.guard_position.get('y') == len(self.map) - 1:
                return True
            elif self.map[self.guard_position.get('y') + 1][self.guard_position.get('x')] == '#':
                self.guard_direction = 'left'
                return False
            else:
                self.guard_position['y'] += 1
                self.map[self.guard_position.get('y')][self.guard_position.get('x')] = 'X'
                return False
        elif self.guard_direction == 'right':
            if self.guard_position.get('x') == len(self.map[0]) - 1:
                return True
            elif self.map[self.guard_position.get('y')][self.guard_position.get('x') + 1] == '#':
                self.guard_direction = 'down'
                return False
            else:
                self.guard_position['x'] += 1
                self.map[self.guard_position.get('y')][self.guard_position.get('x')] = 'X'
                return False
        else:
            if self.guard_position.get('x') == 0:
                return True
            elif self.map[self.guard_position.get('y')][self.guard_position.get('x') - 1] == '#':
                self.guard_direction = 'up'
                return False
            else:
                self.guard_position['x'] -= 1
                self.map[self.guard_position.get('y')][self.guard_position.get('x')] = 'X'
                return False
            
    def _count_distinct(self):
        distinct_sum = 0
        for y, row in enumerate(self.map):
            for x, elem in enumerate(row):
                if elem == 'X':
                    distinct_sum += 1
        print("visited distinct", distinct_sum)
            
    def start_walking(self):
        while True:
            went_out = self._move_guard()
            if went_out:
                break
        self._count_distinct()
        print("loop possibility count", self.loop_possibility_count)
    
# guardTracker = GuardTracker('map.txt')
# guardTracker.start_walking()
```
