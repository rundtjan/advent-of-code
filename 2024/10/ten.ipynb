```python
filename = 'big_map.txt'

with open(filename) as f:
    map = f.readlines()
    map = [x[:len(x) - 1] for x in [list(x) for x in map]]

map
```

```python
zeros = []
for arr in map:
    temp = []
    for elem in map[0]:
        temp.append(0)
    zeros.append(temp[:])
zeros
```

```python
levels = list(range(10))
levels.reverse()

for y in range(len(map)):
    for x in range(len(map[0])):
        if map[y][x] == '9':
            zeros[y][x] = 1
            
            
def inside_of_map(y, x):
    if 0 <= x <= len(map[0]) - 1 and 0 <= y <= len(map) - 1:
        return True
    return False


def spread_around(y, x, val, level):
    for py in [-1, 1]:
        if inside_of_map(y+py, x) and int(map[y+py][x]) == level - 1:
            old_val = zeros[y+py][x]
            zeros[y+py][x] = old_val + val
    for px in [-1, 1]:
        if inside_of_map(y, x+px) and int(map[y][x+px]) == level - 1:
            old_val = zeros[y][x+px]
            zeros[y][x+px] = old_val + val


def draw_zeros():
    print('-----------------------------')
    for row in zeros:
        print(row)

for level in levels:
    draw_zeros()
    for y in range(len(map)):
        for x in range(len(map[0])):
            if int(map[y][x]) == level:
                own_value = zeros[y][x]
                spread_around(y,x, own_value, level)
                    
zeros
```

```python
sum = 0

for y in range(len(map)):
    for x in range(len(map[0])):
        if map[y][x] == '0':
            sum += zeros[y][x]

sum
```
